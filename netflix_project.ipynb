Based on the output, we have identified missing values in various columns and a significant number of unnecessary columns (from Unnamed: 12 onwards). The next step is to:

```python
# Deleting unwanted columns from dataset
df.drop(df.iloc[:,12:26],axis =1,inplace=True)
```

```python
# Ṛeplace the null values in 'director' , 'cast', 'country', 'date_added', 'rating' and 'duration'
# with appropriate placeholders or imputations
df['director'].fillna('Unknown',inplace=True)
df['cast'].fillna('Unknown',inplace=True)
df['country'].fillna('Unknown',inplace=True)
df['duration'].fillna('Unknown duration',inplace=True)
```

```python
# using forward fill method for the null values in column= date_added 
df['date_added'].fillna(method='ffill',inplace=True)
# fill with mode of the column
df['rating'].fillna(df['rating'].mode()[0],inplace = True)
```

Exploratory Data Analysis
We will look into the following sections:

1-Visualize the distribution of movies and TV shows.

2-Analyze the release year trends.

3-distribution of ratings.

4-Look at the top countries producing content.

```python
# calculate the distribution of movies and TV shows
type_counts = df['type'].value_counts()

# plot the distribution of movies and TV shows
# plot the distribution as a pie chart
plt.figure(figsize=(5,5))
plt.title('Distribution of movies and TV shows on Netflix')
colors = ['Red','Blue']
plt.pie(type_counts ,labels = type_counts.index,colors = colors,shadow = True,autopct = '%1.2f%%')
plt.show()

```

```python
# calculate the number of release per year 
release_year_counts = df['release_year'].value_counts().sort_index()

# plot the release year trends
plt.figure(figsize =(10,8))
plt.plot(release_year_counts.index,release_year_counts.values,'bo',linestyle='-')
plt.title('Number of release per year on netflix')
plt.xlabel('Release year')
plt.ylabel('Number of release ')
plt.grid(True)
plt.show()
```

```python
# Distribution of ratings 
rating_counts=df['rating'].value_counts().sort_index()

#plot the distribution as bar chart 
plt.figure(figsize=(12,6))
plt.bar(rating_counts.index,rating_counts.values,)
plt.title('Distribution of ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

```

```python
# calculate the top countries producing content
top_countries =df['country'].value_counts().head(10)

plt.figure(figsize=(10,8))
plt.pie(top_countries,labels=top_countries.index,startangle=140,autopct = "%1.1f%%",colors =plt.cm.Paired.colors)
plt.title("Top 100 Countries producing content on Netflix",y=1.1)
# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis("equal")
plt.show()



```

Number of Content Additiona Over Time on Netflix

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract the year and month from date_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.to_period('M')

# Calculate the number of additions per month
content_additions_per_month = df['month_added'].value_counts().sort_index()

# Plot the number of content additions over time
plt.figure(figsize=(14, 7))
content_additions_per_month.plot(kind='line', marker='o', linestyle='-')
plt.title('Number of Content Additions Over Time on Netflix')
plt.xlabel('Date Added')
plt.ylabel('Number of Additions')
plt.grid(True)
plt.show()
```
